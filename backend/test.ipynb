```python
def str_to_list(str):
    if not(str):
        str=None
    else:
        str=str.strip('[')
        str=str.strip(']')
        str=str.replace("'",'')
        str = str.replace("\t",'')
        str=str.split(', ',str.count(','))
    return str
```
